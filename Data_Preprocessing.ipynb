```python
#we notice than unusual high  and low value are 10800 ,20 respectively 
#we try to use 2 tackled methods in this course  the 3 sigma rule and 1.5 iqr method 
import pandas as pd 
import numpy as np
data = pd.read_csv(r'C:\Users\Admin\Downloads\SPARTA\Capstones\Data Preprocessing')
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')

mean_price = data['Price'].mean()
std_price = data['Price'].std()

# Define the lower and upper bounds for outlier detection using 3-sigma rule
ll = mean_price - 3 * std_price
ul = mean_price + 3 * std_price

# Identify outliers using the 3-sigma rule
outliers_3sigma = data[(data['Price'] < ll) | (data['Price'] > ul)]
outliers_3sigma

```

```python
# And now we try the 1.5 iqr method
import pandas as pd 
import numpy as np
data = pd.read_csv(r'C:\Users\Admin\Downloads\SPARTA\Capstones\Data Preprocessing')
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')

Q1 = data['Price'].quantile(0.25)
Q3 = data['Price'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outlier detection using 1.5 IQR method
lower_bound_iqr = Q1 - 1.5 * IQR
upper_bound_iqr = Q3 + 1.5 * IQR

# Identify outliers using the 1.5 IQR method
outliers_iqr = data[(data['Price'] < lower_bound_iqr) | (data['Price'] > upper_bound_iqr)]
outliers_iqr
```

```python
#also we notice that there is unusual  values that unacceptable for analysis
#for this activity i will use iqr method in lotarea and sale Price column
import pandas as pd
data = pd.read_csv(r'C:\Users\Admin\Downloads\SPARTA\Capstones\Data Preprocessing\property_sales_Part_3_Application.csv')


# Calculate the IQR for SalePrice and LotArea
Q1 = data[['SalePrice', 'LotArea']].quantile(0.25)
Q3 = data[['SalePrice', 'LotArea']].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers in the SalePrice and LotArea columns
outliers = ((data[['SalePrice', 'LotArea']] < lower_bound) | (data[['SalePrice', 'LotArea']] > upper_bound)).any(axis=1)

# Print the rows with outliers
outlier_rows = data[outliers]
print(outlier_rows)
```
